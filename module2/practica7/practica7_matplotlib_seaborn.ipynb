## Задание №1

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_graph():
    x = np.linspace(-10, 10, 100)
    y = (x**2 - 4*np.abs(x) + 3) / (9 - 3*x)
    plt.plot(x, y)
    plt.title('График функции y=(x**2-4*abs(x)+3)/(9-3*x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
plot_graph()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_graph(x0, v0, a):
    t = np.arange(0, 5, 0.1)
    x = (x0+v0+(a*(t**2)/2))
    plt.plot(t, x)
    plt.title('График функции  x = (x0+v0+(a*(t**2)/2))')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
plot_graph(0,10,1)
```

## Задание №3

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Company': ['Yandex', 'Yandex', 'Yandex', 'Yandex'],
        'Year': [2015, 2016, 2017, 2018],
        'Profit': [10, 7, 40, 2]}
df = pd.DataFrame(data)

sns.barplot(x='Year', y='Profit', data=df)
plt.title('Profit for Yandex')

plt.xlabel('Year')
plt.ylabel('Profit')
plt.show()
```

## Задание №4

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
        'Year': [2015, 2016, 2017, 2018],
        'Yandex': [10, 7, 40, 2],
        'Sber': [6, 9, 15, 20]}
df = pd.DataFrame(data)

df_melted = df.melt('Year', var_name='Company', value_name='Profit')
sns.barplot(x='Year', y='Profit', hue='Company', data=df_melted)
plt.title('Profit by Company')

plt.xlabel('Year')
plt.ylabel('Profit')
plt.show()
```

## Задание №5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({'Answer': np.random.choice(['yes', 'no', 'dont know'], size=1000)})

answer_counts = df['Answer'].value_counts()
plt.pie(answer_counts, labels=answer_counts.index, autopct='%1.1f%%')
plt.title('Результаты опроса')
plt.legend(loc='upper left')
plt.show()
```
